```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import os
```

```python
#Load in all the data

snapshot0=22
data0 = np.load('/Users/oonaghparker/Downloads/DISSO/rdr_000{0}_ns_large.part_partid_clumpid.npy'.format(snapshot0))
snapshot1=85
data1 = np.load('/Users/oonaghparker/Downloads/DISSO/rdr_000{0}_ns_large.part_partid_clumpid.npy'.format(snapshot1))
```

```python
# The times are in snap_myr_z_pos. The first column is the snapshot number and the second column is the lookback
# time in Myr
times = np.genfromtxt('/Users/oonaghparker/Downloads/DISSO/snap_myr_z_pos')
```

```python
#Follow a single clump and show it at a certain point in time and plot where it is today (snapshot 85)
#Loops over all the clumpIDs for the snapshot so all figures can be produced in one go
for i in range(0,81):
    cid1=i
    mask0 = np.where(data0['clumpID']==cid1)
    mask1 = np.where(np.isin(data1['ID'],data0['ID'][mask0])) #this line asks what rows in data1 correspond to the IDs from data0

    #Find the Milky Way centre for both snapshots
    mask_mw0 = np.where(data0['clumpID']==0)
    xMW0 = np.median(data0['x'][mask_mw0])
    yMW0 = np.median(data0['y'][mask_mw0])
    zMW0 = np.median(data0['z'][mask_mw0])

    mask_mw1 = np.where(data1['clumpID']==0)
    xMW1 = np.median(data1['x'][mask_mw1])
    yMW1 = np.median(data1['y'][mask_mw1])
    zMW1 = np.median(data1['z'][mask_mw1])

    #Creates a 2x2 figure to plot data
    bins0 = (np.arange(-250,250,3),np.arange(-250,250,3))
    fig,ax = plt.subplots(2,2,figsize=(12,12),sharex=True)

    #ax[0].hist2d(data_snapshot['x'][mask]-xMW,data_snapshot['y'][mask]-yMW,weights=data_snapshot['mass'][mask],bins=bins0,cmap='Greys',norm=LogNorm());
    #ax[1].hist2d(data_snapshot['x'][mask]-xMW,data_snapshot['z'][mask]-zMW,weights=data_snapshot['mass'][mask],bins=bins0,cmap='Greys',norm=LogNorm());

    #Plot snapshot0 data
    ax[0][0].scatter(data0['x'][mask0]-xMW0,data0['y'][mask0]-yMW0,c='r',s=1,rasterized=True,alpha=0.1)
    ax[0][1].scatter(data0['x'][mask0]-xMW0,data0['z'][mask0]-zMW0,c='r',s=1,rasterized=True,alpha=0.1)

    #Plot data from snapshot 85 (present day)
    ax[1][0].scatter(data1['x'][mask1]-xMW1,data1['y'][mask1]-yMW1,c='r',s=1,rasterized=True,alpha=0.1)
    ax[1][1].scatter(data1['x'][mask1]-xMW1,data1['z'][mask1]-zMW1,c='r',s=1,rasterized=True,alpha=0.1)

    for j in range(2):
        for k in range(2):
            ax[j][k].tick_params(labelsize=12)
            ax[j][k].set_aspect('equal')
            ax[j][k].set_xlim(-250,250)
            ax[j][k].set_ylim(-250,250)

    ax[0][0].set_ylabel('y (kpc)',fontsize=12)
    ax[0][0].set_xlabel('x (kpc)',fontsize=12)
    ax[0][1].set_ylabel('z (kpc)',fontsize=12)
    ax[0][1].set_xlabel('x (kpc)',fontsize=12)

    ax[1][0].set_ylabel('y (kpc)',fontsize=12)
    ax[1][0].set_xlabel('x (kpc)',fontsize=12)
    ax[1][1].set_ylabel('z (kpc)',fontsize=12)
    ax[1][1].set_xlabel('x (kpc)',fontsize=12)

    #Adds label onto the figure
    mask_t0 = np.where(times[:,0]==snapshot0)
    ax[0][1].text(100,200,'{0}, {1:.2f} Gyr'.format(snapshot0,1.e-3*np.float(times[:,1][mask_t0])),fontsize=12)

    mask_t1 = np.where(times[:,0]==snapshot1)
    ax[1][1].text(100,200,'{0}, {1:.2f} Gyr'.format(snapshot1,1.e-3*np.float(times[:,1][mask_t1])),fontsize=12)

    plt.savefig('/Users/oonaghparker/Desktop/{0}_{1}_{2}.png'.format(snapshot0,cid1,snapshot1))
```
